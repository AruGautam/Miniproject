```python
#import numpy as np
import cv2
import matplotlib.pyplot as plt
face_cascade = cv2.CascadeClassifier('../input/facedetection22/haarcascade_frontalface_default.xml')
  
image = cv2.imread('../input/face99/WhatsApp Image 2022-02-13 at 17.53.53 (1).png')
grayImage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  
faces = face_cascade.detectMultiScale(grayImage , minNeighbors=7 , scaleFactor=1.1)
  
print(type(faces)) #imread returns a numpy array if image path is given 
  
if len(faces) == 0:
    print("No faces found")
  
else:
    print(faces) #array
    print(faces.shape) #dimensions of the array(rows,columns)
    print("Number of faces detected: " + str(faces.shape[0])) # 0 implies rows of the matrix i.e. no of faces
  
    for (x,y,w,h) in faces:
        cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,0),2) #(image, start_point, end_point, colour, border_radius)
  
    cv2.rectangle(image, ((0,image.shape[0] -30)),(300, image.shape[0]), (255,255,255),-1)
    cv2.putText(image, " Number of faces detected: " + str(faces.shape[0]), (0,image.shape[0] -10), cv2.FONT_HERSHEY_SIMPLEX, 0.57,  (0,0,0), 2)
    #(image, text,org(Position of text),font,font_size, colour of text, thickness of text)
    plt.imshow(image) 
```
